# All the plotting code has to be in one cell jupyter notebook

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('../data/tips.csv')
tips.head()
```

### Plotting with pandas in built plotting functions

```python
tips.tip.plot(kind='hist')
#plt.show()
```

```python
cts = tips.smoker.value_counts()
```

```python
cts
```

```python
cts.plot(kind='bar')
```

### plotting with seaborn

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.distplot(tips.total_bill)
```

```python
sns.distplot(tips.total_bill, kde=False)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips) # linear regression plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=True, col='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=True, col='smoker', row='day')
```

### making multiple plots in a windows is called faceting in seaborn and creating one is by facetgrid

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
