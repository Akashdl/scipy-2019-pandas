## Subsetting by boolean conditions

```python
# for verbosity write colon to select all the columns
df.loc[df["country"] == "United States", :] 
```

```python
df.loc[df["country"] == "United States" & df['year']==1982, :] # pandas throws bitwise operation error 
```

```python
# put operands between paranthesis for logical operators to avoid error thrown by pandas
df.loc[(df["country"] == "United States") & (df['year']==1982), :]
```

## in pandas functions use symbols for logical operators

```python
df.loc[(df["country"] and "United States") & (df['year']==1982), :]
```

## Grouped aggregates (to calculate statistics)
### by grouping the dataframes is split into different dataframes according conditions supplies and
### if working on on parallel compute it computes like an MPI program and combines it together at last 
### and returns a single dataframes back
### we can pass any aggregate function to it
### An aggregate function is one which takes bunch of inputs and returns an input

```python
df.groupby(["year"]) # multiple arguments can be passed as a list
```

```python
df.groupby(["year"])['lifeExp']  # selecting a column to perform calculations on
```

```python
df.groupby(["year"])[['lifeExp', 'pop']] # selecting list of columns to perform calculations on
```

```python
df.groupby(["year"])['lifeExp'].mean() # called an mean aggregate
```

```python
# generic way of calling aggregate function is
import numpy as np
df.groupby(["year"])['lifeExp'].agg(np.std)
```

```python
df.groupby(["year", 'continent'])[['lifeExp', 'gdpPercap']].agg(np.mean)  
# returns heirarchical index which is difficult to handle
```

### we can reset it to normal dataframe indexing format by using resest_index()

```python
df.groupby(["year", 'continent'])[['lifeExp', 'gdpPercap']].agg(np.mean).reset_index()
```
